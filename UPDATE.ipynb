```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('students.db')
cur = conn.cursor()
```

```python
def display_as_table(data, headers):
    df = pd.DataFrame(data=data, columns=[i[0] for i in headers])
    return df
```

```python
cur.execute("SELECT * FROM Students")
display_as_table(cur.fetchall(), cur.description)
```

```python
#It is super important to USE WHERE to define WHAT you want to UPDATE or it updates EVERYTHING
# SET is a ket term when UPDAITN stuff
```

```python
cur.execute("UPDATE Students SET Major='Mathematics' WHERE Major= 'Math'")
cur.execute("SELECT * FROM Students")
display_as_table(cur.fetchall(), cur.description)
```

```python
conn.commit()
```

#conn.rollback() is the opposite of conn.commit().......so if you do something that you want to reverse use rollback before you use commit
.

```python
# Update GPA of 1st year Mathematics majors to be NULL

cur.execute("UPDATE Students SET GPA=NULL WHERE Major='Mathematics' AND Year=1")
cur.execute("SELECT * FROM Students")
display_as_table(cur.fetchall(), cur.description)
```

```python
conn.commit()
```
