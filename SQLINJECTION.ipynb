```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('students.db')
cur = conn.cursor()
```

```python
def display_as_table(data, headers):
    df = pd.DataFrame(data=data, columns=[i[0] for i in headers])
    return df
```

```python
cur.execute("SELECT * FROM Students")
display_as_table(cur.fetchall(),cur.description)
```

```python
cur.execute("UPDATE Students SET GPA=3.3 WHERE Name='Luke'")
```

```python
#defenses from injection include using the "?" and the 
#executemany (it knows to put the gpa and name EXPLICITY to put the gpa and values in the right place)"
def update_student_gpa(name, gpa):    
    statement = "UPDATE Students SET GPA=? WHERE NAME= ?"
    cur.executemany(statement, [(gpa, name)])
```

```python
update_student_gpa("Luke", 3.4)
```
